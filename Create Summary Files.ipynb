```python
df = pd.read_csv('./data/income_expenses_FY2015/all_income_expenses_FY2015.csv')
```

```python
df['account group'].value_counts()
```

```python
expenses  = df[df['account group'] == 'Operating Expenses']
labor_regular_time = df[df['account description'] == 'Regular (Straight Time)']
```

```python
d = expenses
gb = ['account group', 'account description']
#gb = ['Description']

s = (d[['amount']+gb]
     .groupby(gb).sum()).sort_values('amount', ascending=False)[:50]
```

```python
s.plot(kind="")
```
